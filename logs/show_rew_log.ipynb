```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
import math
import re
import pylab
from pylab import figure, show, legend
from mpl_toolkits.axes_grid1 import host_subplot

def draw_fig(file_name, host, label):
    fp = open(file_name, 'r')
    train_batch = []
    train_error = []
    for ln in fp:
      # get train_iterations and train_loss
      if 'episode reward' in ln:
        elements = ln.strip().split(' ')
        train_batch.append(float(elements[4]))
        train_error.append(float(elements[11][:-2]))
    fp.close()
    p1, = host.plot(train_batch, train_error, label=label)
    return p1


host = host_subplot(111)
plt.subplots_adjust(right=1.7, top=1.4) # ajust the right boundary of the plot window

# set labels
host.set_xlabel("train step", fontsize=15)
host.set_ylabel("episode reward", fontsize=15)


# draw_fig(file_name="Pong-v0-human_1env_novae_aw0.0_log", host=host, label="aw0.0")
# draw_fig(file_name="Pong-v0-human_1env_novae_aw0.4_log", host=host, label="aw0.4")
# draw_fig(file_name="Pong-v0-human_1env_novae_aw0.6_log", host=host, label="aw0.6")
# draw_fig(file_name="Pong-v0-human_1env_novae_aw1.0_log", host=host, label="pre_aw1.0")
# draw_fig(file_name="Pong-v0-nopre_aw1.0_log", host=host, label="no_pre_aw1.0")
draw_fig(file_name="Pong-v0-human_1env_novae_aw1.0_1e-4_log", host=host, label="pre_aw1.0_bigrl")
draw_fig(file_name="Pong-v0-human_1env_novae_aw0.66_1e-4_log", host=host, label="aw0.66_bigrl")
draw_fig(file_name="../../rl_a3c_pytorch_nodepart/logs/Pong-v0-cotrain_deep_log", host=host, label="a3c")

host.legend(loc="lower right", fontsize=15)
host.set_xlim([-100,810000])
host.set_ylim([-22, 22.0])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.draw()
plt.savefig("show_accu.pdf", bbox_inches='tight')
plt.show()
```
